### QUESTION 3
### 3. Fibonacci Sequence:
Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt
def fibonacci(n):
    if n <= 0:
        return "Incorrect Input"
    series = [0, 1]
    if n > 2:
        for i in range(2, n):
            series.append(series[i-1] + series[i-2])
    return series

n = int(input("Enter num"))
series = fibonacci(n) #this returns array of fibonacci numbers
plt.plot(list(range(1,n+1)),series)
plt.xlabel("Input")  
plt.ylabel("Fibonacci Series")  
plt.title("Fibonacci Graph")
plt.show()
```

### 11. Remove Duplicates
- Write a program to remove duplicate elements from a list while maintaining the order of elements.
- Calculate the mean and median of the unique elements.

```python
listOfNumsDuplicate = [1,2,2,3,7,4,5,5,6,6,7,6,8,9]
frequencyOfNums = {}
for num in listOfNumsDuplicate:
    if num in frequencyOfNums:
        frequencyOfNums[num] = frequencyOfNums[num] + 1
    else:
        frequencyOfNums[num] = 1
listOfNums = list(frequencyOfNums.keys())
mean = sum(listOfNums) / len(listOfNums)
listOfNums.sort()
median = 0
if len(listOfNums) % 2 == 0: 
    median1 = listOfNums[len(listOfNums)//2] 
    median2 = listOfNums[len(listOfNums)//2 - 1] 
    median = (median1 + median2)/2
else: 
    median = listOfNums[len(listOfNums)//2] 
print(mean,median)
```

### 18. Word Count and Frequency:
Write a program to: 
- Read a file and count the number of words in it.
- Find the most frequent word.
- Visualize the word frequencies using a bar chart.

```python
import matplotlib.pyplot as plt
handle = open("words.txt")
counts = dict()
for line in handle:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word,0) + 1
bigcount = None
bigword = None
for word,count in counts.items():
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count
print(bigword)
plt.bar(range(len(counts)), counts.values(), tick_label=counts.keys())
plt.show()
```

### QUESTION 19
### 19. Random Number Generation:
Write a Python program to:
- Generate 100 random integers between 1 and 50.
- Save them in a text file, one per line.
- Read the file and compute the mean of the numbers.
- Visualize the distribution of the numbers using a histogram.

```python
import random
import matplotlib.pyplot as plt
randomNums = []
for i in range(100):
    randomNums.append(random.randint(1,50))
file = open('data2.txt','w+')
for randomNum in randomNums:
    file.write(f'{randomNum}\n')
file.close()
file = open('data2.txt','r')
content = file.read()
numS = content.split('\n')
sumNums = 0
dictNumFreq = {}
for num in numS:
    if len(num) != 0:
        sumNums = sumNums + int(num)
        if num in dictNumFreq:
            dictNumFreq[num] += 1
        else:
            dictNumFreq[num] = 1
print(sumNums/len(randomNums))
plt.hist(list(dictNumFreq.keys()),bins=10)
plt.show()
```
